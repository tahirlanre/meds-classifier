## Create vocabulary

```python
with open('../src/models/.vector_cache/glove.6B.300d.txt') as f:
    words = f.readlines()
```

```python
len(words)
```

```python
word2id = dict()
```

```python
embeddings = []
```

```python
for i, w in enumerate(words):
    w_list = w.split(' ')
    word2id.update({w_list[0]: i})
    embeddings.append([float(j) for j in w_list[1:]])
```
